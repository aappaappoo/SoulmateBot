````python
# è®ºæ–‡è§£æå·¥å…·

def parse_arxiv_url(url: str) -> dict:
    """è§£æarXivè®ºæ–‡URL"""
    import re
    
    # åŒ¹é… arXiv ID
    patterns = [
        r'arxiv\.org/abs/(\d+\.\d+)',
        r'arxiv\.org/pdf/(\d+\.\d+)',
        r'(\d{4}\.\d{4,5})',
    ]
    
    for pattern in patterns:
        match = re.search(pattern, url)
        if match:
            arxiv_id = match.group(1)
            return {
                "source": "arxiv",
                "id": arxiv_id,
                "abs_url": f"https://arxiv.org/abs/{arxiv_id}",
                "pdf_url": f"https://arxiv.org/pdf/{arxiv_id}.pdf",
            }
    
    return {"source": "unknown", "url": url}

# æµ‹è¯•
test_urls = [
    "https://arxiv.org/abs/2303.08774",
    "https://arxiv.org/pdf/2303.08774.pdf",
    "2303.08774",
]

print("ğŸ”— è®ºæ–‡URLè§£ææµ‹è¯•:")
for url in test_urls:
    result = parse_arxiv_url(url)
    print(f"   {url} -> {result.get('id', 'N/A')}")
````

````python
# æ¼«ç”»åˆ†é•œç”Ÿæˆå™¨

class ComicStoryboard:
    """æ¼«ç”»åˆ†é•œç”Ÿæˆå™¨"""
    
    def __init__(self, title: str, summary: str):
        self.title = title
        self.summary = summary
        self.panels = []
    
    def add_panel(self, scene: str, characters: list, action: str, 
                  dialogue: str = None, style: str = "æ—¥å¼æ¼«ç”»"):
        """æ·»åŠ åˆ†é•œ"""
        self.panels.append({
            "number": len(self.panels) + 1,
            "scene": scene,
            "characters": characters,
            "action": action,
            "dialogue": dialogue,
            "style": style,
        })
    
    def to_prompt(self) -> str:
        """è½¬æ¢ä¸ºå›¾åƒç”Ÿæˆæç¤ºè¯"""
        prompts = []
        for panel in self.panels:
            prompt = f"{panel['style']} style, {panel['scene']}, "
            prompt += f"{', '.join(panel['characters'])}, {panel['action']}"
            prompts.append(prompt)
        return prompts
    
    def display(self):
        """æ˜¾ç¤ºåˆ†é•œ"""
        print(f"ğŸ¬ {self.title}")
        print(f"ğŸ“ {self.summary}")
        print("=" * 60)
        
        for panel in self.panels:
            print(f"\nğŸ“¸ ç¬¬{panel['number']}æ ¼")
            print(f"   åœºæ™¯: {panel['scene']}")
            print(f"   è§’è‰²: {', '.join(panel['characters'])}")
            print(f"   åŠ¨ä½œ: {panel['action']}")
            if panel['dialogue']:
                print(f"   å¯¹è¯: \"{panel['dialogue']}\"")
            print(f"   é£æ ¼: {panel['style']}")

# ç¤ºä¾‹ï¼šåˆ›å»ºä¸€ä¸ªç®€å•çš„æ¼«ç”»åˆ†é•œ
comic = ComicStoryboard(
    title="ç¨‹åºå‘˜çš„ä¸€å¤©",
    summary="ä¸€ä¸ªç¨‹åºå‘˜è§£å†³Bugçš„æ—¥å¸¸"
)

comic.add_panel(
    scene="åŠå…¬å®¤ï¼Œç”µè„‘å‰",
    characters=["ç¨‹åºå‘˜å°æ˜"],
    action="ä¸“æ³¨åœ°ç›¯ç€å±å¹•ï¼Œçœ‰å¤´ç´§é”",
    dialogue="è¿™ä¸ªBugåˆ°åº•åœ¨å“ªé‡Œ...",
    style="æ—¥å¼æ¼«ç”»"
)

comic.add_panel(
    scene="åŒä¸€ä¸ªåŠå…¬å®¤",
    characters=["ç¨‹åºå‘˜å°æ˜"],
    action="çªç„¶çœ¼ç›å‘äº®ï¼ŒåŒæ‰‹ä¸¾èµ·",
    dialogue="æ‰¾åˆ°äº†ï¼ï¼ï¼",
    style="æ—¥å¼æ¼«ç”»"
)

comic.add_panel(
    scene="åŠå…¬å®¤å…¨æ™¯",
    characters=["ç¨‹åºå‘˜å°æ˜", "å…¶ä»–åŒäº‹"],
    action="å°æ˜ç˜«ååœ¨æ¤…å­ä¸Šï¼ŒåŒäº‹ä»¬é¼“æŒ",
    dialogue="åªæ˜¯å°‘äº†ä¸€ä¸ªåˆ†å·...",
    style="æ—¥å¼æ¼«ç”»"
)

comic.display()
````

````python
# ç”Ÿæˆå›¾åƒæç¤ºè¯
prompts = comic.to_prompt()

print("ğŸ–¼ï¸ å›¾åƒç”Ÿæˆæç¤ºè¯:")
print("=" * 60)
for i, prompt in enumerate(prompts, 1):
    print(f"\nç¬¬{i}æ ¼:")
    print(f"   {prompt}")
````
